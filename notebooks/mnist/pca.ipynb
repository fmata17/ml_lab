```python
import numpy as np
from shared.utils import load_dataset
df, X_train, X_test, y_train, y_test = load_dataset("mnist")
X, y = df.iloc[:, :-1], df.iloc[:,-1:]
```

```python
# ERASEME TODO WHY DO THEY LOOK THE SAME?
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# sample data
X_sample = X.sample(500)

# original
pca = PCA(n_components=2)
X_2d = pca.fit_transform(X_sample)

# centered version
X_centered_sample = X_sample - X_sample.mean()
Xc_2d = pca.fit_transform(X_centered_sample)

# plot side by side
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.scatter(X_2d[:,0], X_2d[:,1])
plt.title("Original X")

plt.subplot(1,2,2)
plt.scatter(Xc_2d[:,0], Xc_2d[:,1])
plt.title("Centered X (X - μ)")

plt.show()
```
