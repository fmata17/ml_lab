```python
import kagglehub
import pandas as pd
import numpy as np
import torch

# Download latest version
path = kagglehub.dataset_download("uciml/iris")

print("Path to dataset files:", path)
```

```python
iris = pd.read_csv(path + "/Iris.csv")
# iris.head()
# iris.info()
# iris.describe()
iris
```

```python
# convert to numpy arrays
iris = iris.to_numpy()
```

```python
def one_hot_encode(y: np.ndarray):
    classes = np.unique(y)
    # save label as key to have O(1) lookup of one hot values
    class_to_index = {label: idx for idx, label in enumerate(classes)}
    y_int = np.array([class_to_index[label] for label in y])  # new array of one hot values
    # Create a one-hot encoded matrix by selecting rows from the identity matrix using the class indices
    return np.eye(len(classes))[y_int]
```

```python
# separate features and targets
X = iris[:, 1:-1]  # skip first and last column
y = iris[:, -1]  # last column

X = X.astype(float)  # convert from object (str) to floats
y = one_hot_encode(y)
```
